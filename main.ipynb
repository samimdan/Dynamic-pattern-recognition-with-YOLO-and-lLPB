```python
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import MinMaxScaler
from svglib.svglib import svg2rlg
from reportlab.graphics import renderPDF
```

```python
def svg_to_pdf(svg_path, pdf_path):
    drawing = svg2rlg(svg_path)
    renderPDF.drawToFile(drawing, pdf_path)
```

```python
df = pd.read_csv('results.csv')
df.head()
df_complicated = df[['epoch','train/loss','val/loss','metrics/accuracy_top1']]
						 
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(df_complicated)
scaled_df = pd.DataFrame(scaled_df, columns=df_complicated.columns)
fig = px.line(scaled_df, x='epoch', y=['train/loss', 'val/loss', 'metrics/accuracy_top1'],render_mode='svg')
fig.update_layout(legend_title_text='Metrics')
fig.update_xaxes(title_text='Epoch')
fig.update_yaxes(title_text='Values')
fig.update_layout(font_family="Times New Roman",font_size=14)


fig.write_image("scaled_combine.svg")
fig.show()
svg_to_pdf('scaled_combine.svg', 'train4\pdf\scaled_combine.pdf')
```

```python
#smoothing
df_complicated['train/loss'] = df_complicated['train/loss'].ewm(span=10).mean()
df_complicated['val/loss'] = df_complicated['val/loss'].ewm(span=10).mean()
df_complicated['metrics/accuracy_top1'] = df_complicated['metrics/accuracy_top1'].ewm(span=10).mean()
fig = px.line(df_complicated, x='epoch', y=['train/loss', 'val/loss', 'metrics/accuracy_top1'], title='Combine (Smoothed)',render_mode='svg')
fig = px.line(df_complicated, x='epoch', y=['train/loss', 'val/loss', 'metrics/accuracy_top1'], title='Loss vs Val_loss',render_mode='svg')

fig.add_hline(y=1, line_dash="dot")
fig.add_vline(x=79, line_dash="dot")
fig.add_annotation(x=79, y=1,arrowhead=1, showarrow=True, xshift=0,arrowside="start", ax=50,ay=40 ,text="accuracy = 1")
fig.add_annotation(x=79, y=0,arrowhead=1, showarrow=True, xshift=0,arrowside="start", ax=50,ay=-40 ,text="val/lose = 0")
fig.add_annotation(x=79, y=0.5,arrowhead=1, showarrow=True, xshift=0,arrowside="start", ax=-80,ay=0 ,text="epoch = 79")
fig.update_layout(legend_title_text='Metrics')
fig.update_xaxes(title_text='Epoch')
fig.update_yaxes(title_text='Values')
fig.update_layout(font_family="Times New Roman",font_size=14)

fig.write_image("combine.svg")
fig.show()
```
