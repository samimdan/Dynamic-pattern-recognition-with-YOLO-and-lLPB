```python
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('results.csv')
df.head()
df_complicated = df[['epoch','train/loss','val/loss','metrics/accuracy_top1']]
						 
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(df_complicated)
scaled_df = pd.DataFrame(scaled_df, columns=df_complicated.columns)
fig = px.line(scaled_df, x='epoch', y=['train/loss', 'val/loss', 'metrics/accuracy_top1'], title='Loss vs Val_loss')
fig.show()
```

```python
fig = px.line(df_complicated, x='epoch', y=['train/loss', 'val/loss', 'metrics/accuracy_top1'], title='Loss vs Val_loss')
fig.add_hline(y=1, line_dash="dot",annotation_text="accuracy=1", annotation_position="top right")
fig.show()
```

```python
fig=px.line(df_complicated,x='epoch',y='metrics/accuracy_top1',title='Accuracy')
fig.add_vline(x=79, line_dash="dot",annotation_text="epoch=79", annotation_position="bottom right")
fig.show()
```

```python
fig=px.line(df_complicated,x='epoch',y='train/loss',title='Train Loss')
fig.show()
```

```python
fig=px.line(df_complicated,x='epoch',y='val/loss',title='Val Loss')
fig.show()
```

```python
lr_val_scaled = scaler.fit_transform(df[['epoch', 'lr/pg1', 'val/loss']])
lr_val_scaled = pd.DataFrame(lr_val_scaled, columns=['epoch', 'lr/pg1', 'val/loss'])
fig=px.line(lr_val_scaled, x='epoch', y=['lr/pg1', 'val/loss'], title='Learning Rate')
fig.add_vline(x=0.020, line_dash="dot",annotation_text="epoch=2", annotation_position="bottom right")
fig.show()
```

```python
lr_train_scaled = scaler.fit_transform(df[['epoch', 'lr/pg1', 'train/loss']])
lr_train_scaled = pd.DataFrame(lr_train_scaled, columns=['epoch', 'lr/pg1', 'train/loss'])
fig=px.line(lr_train_scaled, x='epoch', y=['lr/pg1', 'train/loss'], title='Learning Rate')
fig.add_vline(x=0.020, line_dash="dot",annotation_text="epoch=2", annotation_position="bottom right")
fig.show()
```
